## 🏥 Task 1 — Basic Probability: Hospital Blood Types

A hospital records blood types for 200 patients:

| Blood Type | A  | B  | AB | O  |
|------------|----|----|----|----|  
| Count      | 70 | 50 | 20 | 60 |

**Questions:**
1. What is the probability that a randomly selected patient has blood type **A**?
2. What is the probability of selecting a patient with blood type **AB or O**?
3. What is the probability of **NOT** having blood type B?

**Steps to follow:**
- Step 1: Identify the sample space (total patients)
- Step 2: Apply P(event) = favorable outcomes / total outcomes
- Step 3: For 'OR': add the two counts, then divide
- Step 4: For 'NOT': use P(not B) = 1 - P(B)

```python
# Task 1 — Fill in the blanks (replace None with your values)

total_patients = 200   # Total number of patients

count_A  = 70
count_B  = 50
count_AB = 20
count_O  = 60

# Q1: P(A)
P_A = count_A / total_patients
print(f"P(A) = {P_A:.4f} = {Fraction(count_A, total_patients)}")

# Q2: P(AB or O)
P_AB_or_O = (count_AB + count_O) / count_AB
print(f"P(AB or O) = {P_AB_or_O:.4f}")

# Q3: P(not B)
P_not_B = 1 - (count_B / total_patients)
print(f"P(not B) = {P_not_B:.4f}")
```

## 🎲 Task 2 — Independent Events: Rolling Dice & Flipping a Coin

A game show contestant rolls a **fair 6-sided die** and flips a **fair coin**.

**Questions:**
1. What is P(rolling a 4 **AND** getting Heads)?
2. What is P(rolling an even number **AND** getting Tails)?
3. Are the die and coin **independent** events? Justify with the multiplication rule.

**Key formula — Multiplication Rule for Independent Events:**
> P(A ∩ B) = P(A) × P(B)  ← only valid when A and B are independent

**Steps:**
- Step 1: Find P(A) and P(B) individually
- Step 2: Multiply them together
- Step 3: Verify by listing the full sample space and counting

```python
# Task 2

# Q1: P(rolling 4) and P(Heads)
P_roll_4   = 1 / 6       # probability of rolling exactly 4
P_heads    = 1 / 2       # probability of Heads

P_4_and_H  = P_roll_4 * P_heads   # joint probability (independent)
print(f"P(4 AND Heads) = {P_4_and_H:.4f}")

# Q2: P(even AND Tails)
even_numbers = [2, 4, 6]          # list even sides of the die
P_even = len(even_numbers) / 6
P_tails = 1/2

P_even_and_T = P_even * P_tails
print(f"P(Even AND Tails) = {P_even_and_T:.4f}")

# Q3: Verify using full sample space
die_faces = [1, 2, 3, 4, 5, 6]
coin_faces = ['H', 'T']
sample_space = [(d, c) for d in die_faces for c in coin_faces]
print(f"\nTotal outcomes in sample space: {len(sample_space)}")
favorable_4H = [(d, c) for d, c in sample_space if d == 4 and c == 'H']
print(f"Favorable outcomes for (4, H): {favorable_4H}")
print(f"P(4 AND H) from sample space = {len(favorable_4H)}/{len(sample_space)} = {len(favorable_4H)/len(sample_space):.4f}")
print("\n✍️  Are die and coin independent? Explain:")
print("   → yes die and coin are independent events because the outcome of the die does not affect the outcome of the coin, and vice versa.")
```

## 🃏 Task 3 — Dependent Events: Drawing Cards Without Replacement

A standard deck has **52 cards** (4 suits × 13 ranks). You draw two cards **without replacement**.

**Questions:**
1. What is P(first card is a King)?
2. Given the first card **was** a King, what is P(second card is also a King)?
3. What is P(both cards are Kings)? — joint probability of dependent events
4. How would your answer to Q3 change if you drew **with replacement**?

**Key formula — Multiplication Rule for Dependent Events:**
> P(A ∩ B) = P(A) × P(B | A)

```python
# Task 3

total_cards = 52
kings_in_deck = 4

# Q1: P(1st card is King)
P_K1 = kings_in_deck / total_cards
print(f"P(1st King) = {P_K1:.4f} = {Fraction(kings_in_deck, total_cards)}")

# Q2: P(2nd card is King | 1st was King)  → deck now has 51 cards, 3 kings
P_K2_given_K1 = (kings_in_deck - 1) / (total_cards -1)
print(f"P(2nd King | 1st King) = {P_K2_given_K1:.4f} = {Fraction(3, 51)}")

# Q3: P(both Kings) WITHOUT replacement
P_both_kings_no_replace = P_K1 * P_K2_given_K1
print(f"P(Both Kings) without replacement = {P_both_kings_no_replace:.6f}")

# Q4: P(both Kings) WITH replacement
P_both_kings_with_replace = (kings_in_deck / total_cards) * (kings_in_deck / total_cards)
print(f"P(Both Kings) with replacement    = {P_both_kings_with_replace:.6f}")

print("\n✍️  Which probability is higher and WHY?")
print("   → Higher probability occurs with replacement because after drawing a King without replacement, there are fewer Kings left in the deck. \nWithout replacement, the events are dependent, which slightly reduces the joint probability.\nWith replacement, the events are independent, so each draw always has 4 Kings out of 52 cards, giving a larger product.")
```

## 🌦️ Task 4 — Conditional Probability: Weather & Traffic

City traffic data over 365 days shows:
- It rained on **90 days**
- There was heavy traffic on **120 days**
- It rained **AND** there was heavy traffic on **72 days**

**Questions:**
1. What is P(Rain)?
2. What is P(Heavy Traffic)?
3. What is P(Heavy Traffic | Rain) — probability of traffic *given* it rains?
4. What is P(Rain | Heavy Traffic)?
5. Are Rain and Heavy Traffic **independent**? Use the independence test.

**Key formula:**
> P(B | A) = P(A ∩ B) / P(A)

**Independence test:**
> Events A and B are independent iff P(A ∩ B) = P(A) × P(B)

```python
# Task 4

total_days = 365
rain_days  = 90
traffic_days = 120
rain_and_traffic = 72

# Q1 & Q2
P_rain    = rain_days / total_days
P_traffic = traffic_days/ total_days
P_rain_and_traffic = rain_and_traffic / total_days

print(f"P(Rain)           = {P_rain:.4f}")
print(f"P(Traffic)        = {P_traffic:.4f}")
print(f"P(Rain ∩ Traffic) = {P_rain_and_traffic:.4f}")

# Q3: P(Traffic | Rain)
P_traffic_given_rain = P_rain_and_traffic / P_rain
print(f"\nP(Traffic | Rain) = {P_traffic_given_rain:.4f}")
print(f"On a rainy day, there is an {P_traffic_given_rain*100} chance of traffic.")
# Q4: P(Rain | Traffic)
P_rain_given_traffic = P_rain_and_traffic / P_traffic
print(f"P(Rain | Traffic) = {P_rain_given_traffic:.4f}")
print(f"On a day with traffic, there is a {P_rain_given_traffic*100} chance that it is raining")

# Q5: Independence test
P_independent_check = P_rain * P_traffic
print(f"\nIndependence test:")
print(f"P(Rain) × P(Traffic)  = {P_independent_check:.4f}")
print(f"P(Rain ∩ Traffic)     = {P_rain_and_traffic:.4f}")
are_independent = abs(P_independent_check - P_rain_and_traffic) < 1e-9
print(f"Are they independent? → {are_independent}, because If two events are independent, their joint probability is  equal to the product of their individual probabilities.")
print("✍️\nWhat does this tell us about rain and traffic?")
print("   → Rain and traffic are positively associated , traffic is more likely on rainy days.This is consistent with intuition: rainy days increase the probability of traffic jams.")
```

## 🏭 Task 5 — Joint Probability Table: Factory Quality Control

A factory makes products on two shifts. Quality control data:

|              | Defective | Non-Defective | Total |
|--------------|-----------|----------------|-------|
| **Morning**  | 30        | 270            | 300   |
| **Evening**  | 50        | 150            | 200   |
| **Total**    | 80        | 420            | 500   |

**Questions:**
1. Build the **joint probability table** (divide every cell by 500)
2. What is P(Defective AND Morning)?
3. What is P(Defective | Morning)? Compare it to P(Defective | Evening).
4. Which shift has **better quality**?

```python
# Task 5 — Joint Probability Table

data = {
    'Defective':     {'Morning': 30, 'Evening': 50},
    'Non-Defective': {'Morning': 270, 'Evening': 150}
}
total = 500

df_counts = pd.DataFrame(data)
print("=== Count Table ===")
print(df_counts)

# Q1: Build joint probability table
df_joint = df_counts / total      # divide by total
print("\n=== Joint Probability Table ===")
print(df_joint.round(4))

# Q2: P(Defective AND Morning)
P_def_morning = df_joint.loc['Morning', 'Defective']
print(f"\nP(Defective ∩ Morning) = {P_def_morning:.4f}")

# Q3: Conditional probabilities
P_morning = df_counts.loc['Morning'].sum() / total   # marginal P(Morning)
P_evening = df_counts.loc['Evening'].sum() / total

P_def_given_morning = P_def_morning / P_morning
P_def_given_evening = df_joint.loc['Evening', 'Defective'] / P_evening

print(f"P(Defective | Morning) = {P_def_given_morning:.4f}")
print(f"P(Defective | Evening) = {P_def_given_evening:.4f}")

# Q4
print("\n✍️  Which shift has better quality and why?")
print("   → Morning shift has better quality because a smaller proportion of items are defective.")
```

## 🎓 Task 6 — Testing Independence: Study Habits & Grades

A survey of 400 students:

|               | Pass | Fail | Total |
|---------------|------|------|-------|
| **Studies**   | 180  | 20   | 200   |
| **No Study**  | 80   | 120  | 200   |
| **Total**     | 260  | 140  | 400   |

**Questions:**
1. Find P(Pass), P(Studies), P(Pass ∩ Studies)
2. Find P(Pass | Studies) and P(Pass | No Study)
3. Test whether **Study Habit** and **Exam Result** are independent
4. What conclusion can you draw about studying?

```python
# Task 6

n = 400

# Fill in the values from the table
studies_pass   = 180
studies_fail   = 20
nostudy_pass   = 80
nostudy_fail   = 120

P_pass    = (studies_pass + nostudy_pass) / n
P_studies = (studies_pass + studies_fail) / n
P_pass_and_studies = studies_pass / n

print(f"P(Pass)              = {P_pass:.4f}")
print(f"P(Studies)           = {P_studies:.4f}")
print(f"P(Pass ∩ Studies)    = {P_pass_and_studies:.4f}")

# Q2: Conditional probabilities
P_pass_given_studies  = studies_pass / (studies_pass + studies_fail)
P_pass_given_nostudy  = nostudy_pass / (nostudy_pass + nostudy_fail)
print(f"\nP(Pass | Studies)    = {P_pass_given_studies:.4f}")
print(f"P(Pass | No Study)   = {P_pass_given_nostudy:.4f}")

# Q3: Independence test  → if P(Pass ∩ Studies) == P(Pass) × P(Studies)
expected_if_independent = P_pass * P_studies
print(f"\nIf independent: P(Pass)×P(Studies) = {expected_if_independent:.4f}")
print(f"Actual P(Pass ∩ Studies)           = {P_pass_and_studies:.4f}")
print(f"Independent? → {abs(expected_if_independent - P_pass_and_studies) < 1e-9}")

# Visualize
labels = ['Studies\nPass', 'Studies\nFail', 'No Study\nPass', 'No Study\nFail']
values = [studies_pass, studies_fail, nostudy_pass, nostudy_fail]
colors = ['#4CAF50','#f44336','#4CAF50','#f44336']
plt.bar(labels, values, color=colors, edgecolor='black')
plt.title('Study Habits vs Exam Results')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()

print("✍️  Conclusion about studying:")
print("   → Students who study have a much higher chance of passing (0.9 vs 0.4). Studying improves exam success; the events “studying” and “passing” are dependent.")
```

## 📧 Task 7 — Bayes' Theorem: Spam Email Filter

An email system classifies messages. Known statistics:
- **40%** of incoming emails are spam: P(Spam) = 0.40
- The filter flags spam correctly **95%** of the time: P(Flagged | Spam) = 0.95
- The filter **incorrectly** flags non-spam **10%** of the time: P(Flagged | Not Spam) = 0.10

**Question:** If an email is flagged, what is the probability it is actually spam?

**Bayes' Theorem:**
> P(A | B) = [P(B | A) × P(A)] / P(B)
>
> P(B) = P(B|A)·P(A) + P(B|A')·P(A') ← Total Probability

**Steps:**
1. Define all given probabilities
2. Calculate P(Flagged) using the law of total probability
3. Apply Bayes' formula

```python
import matplotlib.pyplot as plt

# Task 7 — Bayes' Theorem: Spam Filter

# Step 1: Define given probabilities
P_spam         = 0.40   # P(Spam)
P_not_spam     = 1 - P_spam

P_flag_spam    = 0.95   # P(Flagged | Spam)
P_flag_notspam = 0.10   # P(Flagged | Not Spam)

# Step 2: Total probability of being flagged — P(Flagged)
P_flagged = (P_flag_spam * P_spam) + (P_flag_notspam * P_not_spam)
print(f"P(Flagged) = {P_flag_spam} × {P_spam} + {P_flag_notspam} × {P_not_spam}")
print(f"P(Flagged) = {P_flagged:.4f}")

# Step 3: Bayes' theorem → P(Spam | Flagged)
P_spam_given_flagged = (P_flag_spam * P_spam) / P_flagged
print(f"\nP(Spam | Flagged) = {P_spam_given_flagged:.4f} ({P_spam_given_flagged*100:.1f}%)")

# Visualize with a simple probability tree
fig, ax = plt.subplots(figsize=(9, 5))
ax.axis('off')
ax.set_xlim(0, 10); ax.set_ylim(0, 10)
ax.text(0.5, 5, 'Email\n(Prior)', fontsize=11, ha='center', va='center',
        bbox=dict(boxstyle='round', facecolor='#bbdefb'))
ax.annotate('', xy=(3.5, 7.5), xytext=(1.5, 5.5),
            arrowprops=dict(arrowstyle='->', color='green', lw=2))
ax.text(3.8, 7.5, f'Spam\nP={P_spam}', fontsize=10, va='center',
        bbox=dict(boxstyle='round', facecolor='#ffccbc'))
ax.annotate('', xy=(3.5, 2.5), xytext=(1.5, 4.5),
            arrowprops=dict(arrowstyle='->', color='red', lw=2))
ax.text(3.8, 2.5, f'Not Spam\nP={P_not_spam}', fontsize=10, va='center',
        bbox=dict(boxstyle='round', facecolor='#c8e6c9'))
ax.annotate('', xy=(7.5, 8.5), xytext=(5.5, 8.0),
            arrowprops=dict(arrowstyle='->', color='green', lw=1.5))
ax.text(7.6, 8.5, f'Flagged  P={P_flag_spam}', fontsize=9)
ax.annotate('', xy=(7.5, 6.8), xytext=(5.5, 7.2),
            arrowprops=dict(arrowstyle='->', color='gray', lw=1.5))
ax.text(7.6, 6.8, f'Not Flagged  P={1-P_flag_spam}', fontsize=9)
ax.annotate('', xy=(7.5, 3.2), xytext=(5.5, 2.8),
            arrowprops=dict(arrowstyle='->', color='red', lw=1.5))
ax.text(7.6, 3.2, f'Flagged  P={P_flag_notspam}', fontsize=9)
ax.annotate('', xy=(7.5, 1.5), xytext=(5.5, 2.2),
            arrowprops=dict(arrowstyle='->', color='gray', lw=1.5))
ax.text(7.6, 1.5, f'Not Flagged  P={1-P_flag_notspam}', fontsize=9)
ax.set_title('Probability Tree: Spam Filter', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n✍️  Interpret the result: what does this mean for the email system?")
print("   → If the system flags an email, there is an 86.4% chance it is actually spam. "
      "While the filter is effective, about 13.6% of flagged emails are still legitimate (False Positives).")
```

## 🦠 Task 8 — Bayes' Theorem: Medical Diagnosis

A rare disease affects **1%** of the population. A medical test has:
- **True positive rate** (sensitivity): P(Test+ | Disease) = 0.99
- **False positive rate**: P(Test+ | No Disease) = 0.05

**Question:** If a patient tests **positive**, what is the probability they actually have the disease?

**Steps:**
1. Set up prior probabilities P(Disease) and P(No Disease)
2. Compute P(Test+) using total probability
3. Apply Bayes' theorem
4. Interpret — is the result surprising?

```python
# Task 8 — Medical Test

# Step 1: Prior probabilities
P_disease     = 0.01   # 1% of population
P_no_disease  = 1 - P_disease

# Likelihoods
P_pos_given_disease     = 0.99   # sensitivity = 0.99
P_pos_given_no_disease  = 0.05  # false positive rate = 0.05

# Step 2: P(Test+) — total probability
P_test_positive = (P_pos_given_disease * P_disease) + \
                  (P_pos_given_no_disease * P_no_disease)
print(f"P(Test+) = {P_test_positive:.4f}")

# Step 3: Bayes' theorem
P_disease_given_pos = (P_pos_given_disease * P_disease) / P_test_positive
print(f"P(Disease | Test+) = {P_disease_given_pos:.4f}  ({P_disease_given_pos*100:.2f}%)")

# Visualise the 'base rate neglect' effect
population = 10000
has_disease = int(population * P_disease)
no_disease  = population - has_disease

TP = int(has_disease * P_pos_given_disease)       # true positives
FP = int(no_disease  * P_pos_given_no_disease)    # false positives
total_pos = TP + FP

print(f"\nSimulation with {population} people:")
print(f"  Have disease:         {has_disease}")
print(f"  True positives (TP):  {TP}")
print(f"  False positives (FP): {FP}")
print(f"  P(Disease | +) ≈ {TP}/{total_pos} = {TP/total_pos:.4f}")

plt.pie([TP, FP], labels=[f'True Positive\n({TP})', f'False Positive\n({FP})'],
        colors=['#ef5350','#ffcc80'], autopct='%1.1f%%', startangle=140)
plt.title(f'Among all positive tests (n={total_pos})', fontsize=12)
plt.show()

print("\n✍️  Why is the result surprisingly low even though the test is 99% accurate?")
print("   → The disease is very rare (1% prevalence).\nMost positive tests are false positives because the majority of the population does not have the disease.\nThis is called the base rate effect or base rate neglect: the low prevalence dominates the probability."
      "\nEven highly accurate tests can give low probability of disease after a positive result if the disease is rare. Always consider prevalence (base rate) when interpreting medical test")
```

## 🎒 Task 9 — Bayes' Theorem: Bag of Marbles (Multi-Cause)

There are **3 bags**:
- **Bag 1**: 3 Red, 7 Blue  
- **Bag 2**: 6 Red, 4 Blue  
- **Bag 3**: 5 Red, 5 Blue  

A bag is chosen at random (equal probability = 1/3), and **one marble is drawn — it is Red**.

**Questions:**
1. What is the prior probability of choosing each bag?
2. What is P(Red | Bag 1), P(Red | Bag 2), P(Red | Bag 3)?
3. Using Bayes' Theorem, find P(Bag 1 | Red), P(Bag 2 | Red), P(Bag 3 | Red)
4. Do the three posterior probabilities sum to 1? (Verify!)

**Steps:**
1. Prior: P(Bag_i) for each bag
2. Likelihood: P(Red | Bag_i)
3. Evidence: P(Red) = Σ P(Red | Bag_i) × P(Bag_i)
4. Posterior: P(Bag_i | Red) = [P(Red | Bag_i) × P(Bag_i)] / P(Red)

```python
# Task 9

bags = {
    'Bag 1': {'red': 33, 'blue': 7},   # fill in
    'Bag 2': {'red': 6, 'blue': 4},
    'Bag 3': {'red': 5, 'blue': 5},
}

# Step 1: Prior probability (equal)
prior = {bag: 1/3 for bag in bags}

# Step 2: Likelihood P(Red | Bag_i)
likelihood = {}
for bag, counts in bags.items():
    total_marbles = counts['red'] + counts['blue']
    likelihood[bag] = counts['red'] / total_marbles

print("Likelihoods P(Red | Bag):")
for bag, p in likelihood.items():
    print(f"  {bag}: {p:.4f}")

# Step 3: Evidence P(Red)
P_red = sum(likelihood[bag] * prior[bag] for bag in bags)
print(f"\nP(Red) = {P_red:.4f}")

# Step 4: Posterior using Bayes
posterior = {}
for bag in bags:
    posterior[bag] = (likelihood[bag] * prior[bag]) / P_red

print("\nPosterior Probabilities P(Bag | Red):")
for bag, p in posterior.items():
    print(f"  {bag}: {p:.4f}")

print(f"\nSum of posteriors: {sum(posterior.values()):.6f}  ← should be 1.0")

# Visualize: Prior vs Posterior
x = np.arange(3)
width = 0.35
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x - width/2, prior.values(),   width, label='Prior',     color='#90caf9')
ax.bar(x + width/2, posterior.values(), width, label='Posterior', color='#ef9a9a')
ax.set_xticks(x); ax.set_xticklabels(bags.keys())
ax.set_ylabel('Probability'); ax.set_title('Prior vs Posterior Probability of Bag')
ax.legend(); ax.set_ylim(0, 0.6)
plt.tight_layout(); plt.show()

print("✍️  Which bag most likely produced the red marble?")
print("   → The bag most likely to have produced the red marble is Bag 2 (posterior ≈ 0.429). \nBag 3 is next likely (≈ 0.357), and Bag 1 is least likely (≈ 0.214).")
```

## 🤖 Task 10 — Putting It All Together: Machine Failure Diagnosis

A factory has **3 machines** producing widgets:
- Machine A produces **50%** of widgets, defect rate **2%**
- Machine B produces **30%** of widgets, defect rate **3%**
- Machine C produces **20%** of widgets, defect rate **5%**

A widget is picked at random and found to be **defective**.

**Questions:**
1. What is the **overall probability** that a random widget is defective — P(Defective)?
2. Given the widget is defective, what is the probability it came from **Machine A**? B? C?
3. Which machine is the **most likely culprit**?
4. If Machine C's defect rate drops to **1%**, how do the posterior probabilities change?

**This task combines:**
- Law of Total Probability (for Q1)
- Bayes' Theorem (for Q2)
- Sensitivity analysis (for Q4)

```python
# Task 10 — Full Bayes Analysis

# --- Define machines: (production_share, defect_rate) ---
machines = {
    'Machine A': {'share': 0.50, 'defect_rate': 0.02},   # 50%, 2%
    'Machine B': {'share': 0.30, 'defect_rate': 0.03},   # 30%, 3%
    'Machine C': {'share': 0.20, 'defect_rate': 0.05},   # 20%, 5%
}

def bayes_machines(machines):
    """Compute P(Defective) and P(Machine | Defective) using Bayes."""
    # Q1: Total probability of defect
    P_defect = sum(m['share'] * m['defect_rate'] for m in machines.values())
    
    # Q2: Posterior for each machine
    posteriors = {}
    for name, m in machines.items():
        posteriors[name] = (m['share'] * m['defect_rate']) / P_defect
    return P_defect, posteriors

P_defect, posteriors = bayes_machines(machines)

print(f"Q1: P(Defective) = {P_defect:.4f} ({P_defect*100:.2f}%)\n")
print("Q2: P(Machine | Defective):")
for name, p in posteriors.items():
    print(f"   {name}: {p:.4f} ({p*100:.1f}%)")

# Q3: Most likely culprit
culprit = max(posteriors, key=posteriors.get)
print(f"\nQ3: Most likely source: {culprit}")

# Q4: Sensitivity — drop Machine C defect rate to 1%
machines_new = dict(machines)   # copy
machines_new['Machine C'] = {'share': 0.20, 'defect_rate': 0.01}

P_defect_new, posteriors_new = bayes_machines(machines_new)
print(f"\nQ4: After improvement — P(Defective) = {P_defect_new:.4f}")
print("Updated posteriors:")
for name, p in posteriors_new.items():
    print(f"   {name}: {p:.4f} ({p*100:.1f}%)")

# Visualise: before vs after
labels = list(posteriors.keys())
before = list(posteriors.values())
after  = list(posteriors_new.values())
x = np.arange(len(labels))
fig, ax = plt.subplots(figsize=(9, 5))
ax.bar(x - 0.2, before, 0.4, label='Before (C=5%)', color='#ef9a9a')
ax.bar(x + 0.2, after,  0.4, label='After (C=1%)',  color='#a5d6a7')
ax.set_xticks(x); ax.set_xticklabels(labels)
ax.set_ylabel('P(Machine | Defective)')
ax.set_title('Posterior Probability — Before vs After Improvement')
ax.legend(); plt.tight_layout(); plt.show()

print("\n✍️  Reflect: How does improving Machine C affect blame on Machine A and B?")
print("   →Reducing Machine's C defect rate greatly lowers its chance of being culprit. Now Machine A is most likely responsible ")
```
