## 📝 Task 7 — List Comprehensions

### 🎯 Your Task
Use **list comprehensions** (no regular for-loops!) to create and print each of the following:

1. A list of the **squares** of numbers 1 through 15: `[1, 4, 9, 16, ...]`
2. A list of numbers from 1–50 that are **divisible by 3 or 5** (but not both): `[3, 5, 6, 9, 10, ...]`
3. A list of the **lengths** of each word in the sentence `"the quick brown fox jumps over the lazy dog"`
4. A list of **uppercase versions** of only the words that have **5 or more letters**: `['QUICK', 'BROWN', 'JUMPS', 'LAZY']` *(hint: 'over' has 4, skip it)*

> 💡 **Hint for #2:** `(n % 3 == 0) != (n % 5 == 0)` is Python's way to express XOR for two conditions.  
> 💡 **Hint for #3:** `sentence.split()` gives you the word list.


````python
# 1. Squares of 1 to 15
# YOUR CODE HERE
square_nums = [x**2 for x in range(1,16) ]
print(f"Square numbers: {square_nums}")

# 2. Numbers 1-50 divisible by 3 OR 5 but NOT both
# YOUR CODE HERE
divisible_3_5 = [x for x in range(1,51) if ((x % 3 == 0) != (x % 5 == 0))]
print(f"Divisible 3 xor 5: {divisible_3_5}")

# 3. Lengths of each word in the sentence
sentence = "the quick brown fox jumps over the lazy dog"
len_word = [len(word) for word in sentence.split()]
print(f"Lengths of each word in the sentence: {len_word}")

# 4. Uppercase words with 5 or more letters
letters_5_more = [word for word in sentence.split() if (len(word) >= 5)]
print(f"Uppercase words with 5 or more letters: {letters_5_more}")
````

### Background
Just like list comprehensions, you can build dictionaries in one line:

Dictionaries are perfect for counting frequencies — a pattern you'll use constantly in data science.

### 🎯 Your Task

**Part A — Dictionary Comprehension:**
1. Create a dictionary mapping each number 1–10 to its **cube**: `{1: 1, 2: 8, 3: 27, ...}`
2. Create a dictionary mapping each number 1–20 to `"even"` or `"odd"`: `{1: 'odd', 2: 'even', ...}`

**Part B — Word Frequency Counter:**  
Count how many times each word appears in the text below, ignoring case. Print the result sorted by **frequency (descending)**.

```python
text = "to be or not to be that is the question to be is to think"
```
Expected top entries: `to: 4, be: 3, is: 2, ...`

> 💡 **Hint for Part B:** Start with `freq = {}`, then loop over `text.lower().split()`.  
> Use `freq[word] = freq.get(word, 0) + 1` to safely increment counts.  
> Sort with `sorted(freq.items(), key=lambda x: x[1], reverse=True)`.


````python
# Part A — Dictionary comprehensions
# 1. Cubes of 1 to 10
# YOUR CODE HERE
cube_dic = {key: key**3 for key in range(1,11)}
print(f"Cube dic: {cube_dic}")

# 2. Even/Odd mapping for 1 to 20
even_odd_dic = {key:("even" if key%2==0 else "odd") for key in range(1,21)}
print(f"Even odd dic: {even_odd_dic}")

# Part B — Word frequency counter
text = "to be or not to be that is the question to be is to think"
freq= {}
for word in text.lower().split():
    if word in freq:
        freq[word] = freq[word] + 1
    else:
        freq[word] = 1
print(f"Frequent words: {freq}")
print(f"sorted frequence:{sorted(freq.items(), key=lambda x:x[1], reverse=True)}")
````

# 🔵 Section 5 — Sets


## 📝 Task 12 — Set Operations

### Background
A **set** is an **unordered collection of unique elements**. Duplicate values are automatically removed. Sets support mathematical operations: union (`|`), intersection (`&`), difference (`-`), symmetric difference (`^`).

| Operation | Symbol | Method | Meaning |
|-----------|--------|--------|---------|
| Union | `\|` | `.union()` | All elements from both sets |
| Intersection | `&` | `.intersection()` | Only elements in **both** |
| Difference | `-` | `.difference()` | Elements in A but **not** B |
| Symmetric Difference | `^` | `.symmetric_difference()` | Elements in either but **not both** |

### 🎯 Your Task
```python
python_students = {"Alice", "Bob", "Carol", "David", "Eve"}
java_students   = {"Bob", "David", "Frank", "Grace", "Alice"}
```

1. Find students enrolled in **both** courses (intersection)
2. Find students enrolled in **at least one** course (union)
3. Find students taking **Python but NOT Java** (difference)
4. Find students taking **exactly one** course (symmetric difference)
5. Check if `{"Bob", "Alice"}` is a **subset** of `python_students`
6. Add `"Henry"` to `python_students` and remove `"Carol"` from it
7. Create a list `[1, 2, 2, 3, 3, 3, 4, 4, 4, 4]` and use a set to print only the **unique values**

> 💡 **Hint:** Use `.issubset()` or the `<=` operator for subset check.


````python
python_students = {"Alice", "Bob", "Carol", "David", "Eve"}
java_students   = {"Bob", "David", "Frank", "Grace", "Alice"}

# YOUR CODE HERE
# 1. Intersection
print(f"Students must enrolled in both course:{python_students & java_students}")

# 2. Union
print(f"Students at least in one course course:{python_students | java_students}")


# 3. Difference (Python but NOT Java)
print(f"Students taking Python but NOT Java:{python_students - java_students}")


# 4. Symmetric difference
print(f"Students excatly one course :{python_students ^ java_students}")

# 5. Subset check: {"Bob", "Alice"} ⊆ python_students
newset = {"Bob", "Alice"}
print(f"Subset check :{newset <= python_students}")
# print(f"Subset check :{newset.issubset(python_students)}")


# 6. Add "Henry", remove "Carol"
python_students.add("henry")
python_students.discard("Carol")
print("Upsated set by adding henry and removing carol in python",python_students)
# 7. Unique values from list
nums = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
print(set(nums))
````

## 📝 Task 18 — Capstone: Combined Data Structures Challenge

### Background
Real data science tasks require combining **all** the data structures you have learned. This final task tests your ability to choose the right structure for each job.

### 🎯 Your Task
You are given student records as a **list of dictionaries**:

```python
students = [
    {"name": "Alice",   "scores": [85, 90, 78], "city": "Lahore"},
    {"name": "Bob",     "scores": [60, 55, 70], "city": "Karachi"},
    {"name": "Carol",   "scores": [92, 88, 95], "city": "Islamabad"},
    {"name": "David",   "scores": [70, 65, 80], "city": "Lahore"},
    {"name": "Eve",     "scores": [50, 45, 55], "city": "Karachi"},
    {"name": "Frank",   "scores": [88, 92, 84], "city": "Islamabad"},
    {"name": "Grace",   "scores": [75, 80, 78], "city": "Lahore"},
]
```

Complete each of the following:

1. **Average score:** Add a key `"average"` to each student dict containing the average of their `scores` (round to 2 decimal places). Print all names and averages.

2. **Grade assignment:** Using a dictionary comprehension, create a dict `{name: grade}` where grade is:
   - `"A"` if average ≥ 85
   - `"B"` if average ≥ 70
   - `"C"` if average ≥ 55
   - `"F"` otherwise

3. **Top students:** Use a list comprehension to create a list of **names** of students with average ≥ 80.

4. **Unique cities:** Use a **set** to find all unique cities, then print them.

5. **NumPy statistics:** Convert all scores into a **2D NumPy array** (shape `7×3`). Compute and print:
   - Overall mean, min, max
   - Mean score per student (axis=1)
   - Mean score per test/column (axis=0)

6. **Pandas summary:** Create a **DataFrame** with columns `Name`, `Average`, `Grade`, `City`. Print it, then print the **average score per city** using `.groupby()`.

> 💡 **Hints:**
> - `sum(scores) / len(scores)` or `np.mean(scores)` for averages
> - `np.array([s["scores"] for s in students])` to build the 2D array
> - For the DataFrame, build a list of dicts or use parallel lists


````python
import numpy as np
import pandas as pd

students = [
    {"name": "Alice",  "scores": [85, 90, 78], "city": "Lahore"},
    {"name": "Bob",    "scores": [60, 55, 70], "city": "Karachi"},
    {"name": "Carol",  "scores": [92, 88, 95], "city": "Islamabad"},
    {"name": "David",  "scores": [70, 65, 80], "city": "Lahore"},
    {"name": "Eve",    "scores": [50, 45, 55], "city": "Karachi"},
    {"name": "Frank",  "scores": [88, 92, 84], "city": "Islamabad"},
    {"name": "Grace",  "scores": [75, 80, 78], "city": "Lahore"},
]

# Step 1: Add "average" key to each student dict
for student in students:
    student['average'] = sum(student['scores']) / len(student['scores'])

print("Step 1 — Students with averages:")
for s in students:
    print(s)
print()

# Step 2: Dict comprehension {name: grade} — assign letter grades
# Example: >=90:A, 80-89:B, 70-79:C, 60-69:D, <60:F
grade_dict = {s['name']: ('A' if s['average']>=90 else
                           'B' if s['average']>=80 else
                           'C' if s['average']>=70 else
                           'D' if s['average']>=60 else
                           'F')
              for s in students}

print("Step 2 — Name: Grade")
print(grade_dict)
print()

# Step 3: List of names with average >= 80
top_students = [s['name'] for s in students if s['average'] >= 80]
print("Step 3 — Students with average >=80:", top_students)
print()

# Step 4: Set of unique cities
cities = {s['city'] for s in students}
print("Step 4 — Unique cities:", cities)
print()

# Step 5: NumPy 2D array and statistics
scores_array = np.array([s['scores'] for s in students])
print("Step 5 — NumPy scores array:\n", scores_array)
print("Mean per student:", scores_array.mean(axis=1))
print("Mean per subject:", scores_array.mean(axis=0))
print("Overall mean:", scores_array.mean())
print()

# Step 6: Pandas DataFrame and groupby city
df = pd.DataFrame(students)
print("Step 6 — DataFrame:\n", df)
print("\nAverage per city:\n", df.groupby('city')['average'].mean())
````
